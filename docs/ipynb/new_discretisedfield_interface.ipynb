What should happen to the value. Is field represented with value and norm, or just value?

```python
field.value
```

```python
field.value = (10, 10, 10)
```

Norm remains the same.

```python
field.value
```

```python
field.norm
```

Norm is a Python function. For instance, space varying Ms.

```python
def norm(pos):
    x, y, z = pos
    return x+y+z

field.norm = norm
```

Norm does not have a unique representation any more.

```python
field.norm
```
